```python
df_aircrashes['Year'] = df_aircrashes['Date'].dt.year
df_aircrashes['Month'] = df_aircrashes['Date'].dt.strftime('%B')
```

```python
months = [
    "January", "February", "March", "April", "May", "June",
    "July", "August", "September", "October", "November", "December"
]

df_airtraffic['Month'] = df_airtraffic['Month'].apply(lambda x: months[x - 1])
df_airtraffic['Month']
```

```python
df_merged_set = pd.merge(df_airtraffic, df_aircrashes, how='outer', on=['Year', 'Month'])
```

```python
df_merged_set.dropna(subset=['Fatalities'], inplace=True)
df_merged_set['Published Airline IATA Code'].fillna(df_merged_set['Published Airline IATA Code'].mode()[0])
df_merged_set['Operating Airline IATA Code'].fillna(df_merged_set['Operating Airline IATA Code'].mode()[0])
df_merged_set['Aboard'].fillna(df_merged_set['Aboard'].mean())
df_merged_set['Fatalities'].fillna(df_merged_set['Fatalities'].mean())
df_merged_set['Ground'].fillna(df_merged_set['Ground'].mean())
df_merged_set['cn/In'].fillna(df_merged_set['cn/In'].mode()[0])
df_merged_set['Registration'].fillna(df_merged_set['Registration'].mode()[0])
df_merged_set['Type'].fillna(df_merged_set['Type'].mode()[0])
df_merged_set['Route'].fillna(df_merged_set['Route'].mode()[0])
df_merged_set['Operator'].fillna(df_merged_set['Operator'].mode()[0])
df_merged_set['Location'].fillna(df_merged_set['Location'].mode()[0])
df_merged_set['Time'].fillna(df_merged_set['Time'].mode()[0])
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x='Operator', y='Fatalities', data=df_merged_set)
plt.xticks(rotation=90)
plt.show()
```

```python
df_merged_set['Time'] = df_merged_set['Time'].fillna('00:00:00')

df_merged_set['Aboard'].fillna(df_merged_set['Aboard'].mean())
df_merged_set['Ground'].fillna(df_merged_set['Ground'].mean())

df_merged_set.isnull().sum()

df_merged_set['Aboard'] = pd.to_numeric(df_merged_set['Aboard'], errors='coerce')
df_merged_set['Ground'] = pd.to_numeric(df_merged_set['Ground'], errors='coerce')
df_merged_set[['Aboard', 'Ground']].dtypes
df_merged_set['Aboard'] = df_merged_set['Aboard'].fillna(df_merged_set['Aboard'].mean())
df_merged_set['Ground'] = df_merged_set['Ground'].fillna(df_merged_set['Ground'].mean())

df_merged_set[['Aboard', 'Ground']].isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_merged_set['Operator'] = le.fit_transform(df_merged_set['Operator'])
df_merged_set['Operating Airline'] = le.fit_transform(df_merged_set['Operating Airline'])
df_merged_set['Operating Airline IATA Code'] = le.fit_transform(df_merged_set['Operating Airline IATA Code'])
df_merged_set['Published Airline'] = le.fit_transform(df_merged_set['Published Airline'])
df_merged_set['Published Airline IATA Code'] = le.fit_transform(df_merged_set['Published Airline IATA Code'])
df_merged_set['GEO Summary'] = le.fit_transform(df_merged_set['GEO Summary'])
df_merged_set['GEO Region'] = le.fit_transform(df_merged_set['GEO Region'])
df_merged_set['Activity Type Code'] = le.fit_transform(df_merged_set['Activity Type Code'])
df_merged_set['Price Category Code'] = le.fit_transform(df_merged_set['Price Category Code'])
df_merged_set['Terminal'] = le.fit_transform(df_merged_set['Terminal'])
df_merged_set['Boarding Area'] = le.fit_transform(df_merged_set['Boarding Area'])
df_merged_set['Adjusted Activity Type Code'] = le.fit_transform(df_merged_set['Adjusted Activity Type Code'])
df_merged_set['Month'] = le.fit_transform(df_merged_set['Month'])
df_merged_set['Location'] = le.fit_transform(df_merged_set['Location'])
df_merged_set['Flight #'] = le.fit_transform(df_merged_set['Flight #'])
df_merged_set['Route'] = le.fit_transform(df_merged_set['Route'])
df_merged_set['Type'] = le.fit_transform(df_merged_set['Type'])
df_merged_set['Registration'] = le.fit_transform(df_merged_set['Registration'])
df_merged_set['cn/In'] = le.fit_transform(df_merged_set['cn/In'])
df_merged_set['Summary'] = le.fit_transform(df_merged_set['Summary'])

```

```python
from sklearn.model_selection import train_test_split

x = df_merged_set.drop('Fatalities', axis=1)
y = df_merged_set['Fatalities']

X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print(X_train.isna().sum())
print(Y_train.isna().sum()) 
```

```python

X_train['Activity Period'] = pd.to_numeric(X_train['Activity Period'], errors='coerce')
X_train['Passenger Count'] = pd.to_numeric(X_train['Passenger Count'], errors='coerce')
X_train['Adjusted Passenger Count'] = pd.to_numeric(X_train['Adjusted Passenger Count'], errors='coerce')

X_train['Date'] = pd.to_datetime(X_train['Date'], format='%m/%d/%Y', errors='coerce')
X_train['Date'] = X_train['Date'].astype('int64') / 10**9

X_train['Date'].isnull().sum()

X_train['Passenger Count'] = pd.to_numeric(X_train['Passenger Count'], errors='coerce')
X_train['Passenger Count'] = X_train['Passenger Count'].fillna(X_train['Passenger Count'].mean())


X_train['Adjusted Passenger Count'] = pd.to_numeric(X_train['Adjusted Passenger Count'], errors='coerce')
X_train['Adjusted Passenger Count'] = X_train['Adjusted Passenger Count'].fillna(X_train['Adjusted Passenger Count'].mean())
X_train['Adjusted Passenger Count'].isnull().sum()

X_train['Activity Period'] = pd.to_numeric(X_train['Activity Period'], errors='coerce')
X_train['Activity Period'] = X_train['Activity Period'].fillna(X_train['Activity Period'].median())
X_train['Activity Period'].isnull().sum()

X_train['Time'] = pd.to_datetime(X_train['Time'], format='%H:%M:%S', errors='coerce').dt.time

X_train.dtypes
```

```python
X_train['Activity Period'] = pd.to_datetime(X_train['Activity Period'], errors='coerce')

X_train['Activity Period'] = X_train['Activity Period'].apply(lambda x: x.timestamp() if pd.notna(x) else 0)

l_encoder = LabelEncoder()
Y_train = l_encoder.fit_transform(Y_train)
X_train.drop('Time', axis=1)
#print(X_train.dtypes)
```

```python
Y_train = (Y_train > 20).astype(int)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

logistic_model = LogisticRegression(max_iter=1000)
logistic_model.fit(X_train, Y_train)
```

```python
Y_pred = logistic_model.predict(X_train)
```

```python
accuracy = accuracy_score(Y_train, Y_pred)
print("Accuracy:", accuracy)

print("Confusion Matrix:")
print(confusion_matrix(Y_train, Y_pred))

print("Classification Report:")
print(classification_report(Y_train, Y_pred))
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(logistic_model, X_train, Y_train, cv=5)
print("Cross-validation scores:", scores)
print("Mean cross-validation score:", scores.mean())
```

```python
#print("X_train columns: ", X_train.columns)


X_test = X_test[X_train.columns]

print("X_test columns: ", X_test.columns)
```

```python
X_test['Date'] = pd.to_datetime(X_test['Date'], format='%m/%d/%Y', errors='coerce')
X_test['Date'] = X_test['Date'].astype('int64') / 10**9

X_test['Date'].dtype

X_test.isnull().sum()
```

```python
X_test['Passenger Count'] = pd.to_numeric(X_test['Passenger Count'], errors='coerce')
X_test['Passenger Count'] = X_test['Passenger Count'].fillna(X_test['Passenger Count'].mean())


X_test['Adjusted Passenger Count'] = pd.to_numeric(X_test['Adjusted Passenger Count'], errors='coerce')
X_test['Adjusted Passenger Count'] = X_test['Adjusted Passenger Count'].fillna(X_test['Adjusted Passenger Count'].mean())

X_test['Activity Period'] = pd.to_numeric(X_test['Activity Period'], errors='coerce')
X_test['Activity Period'] = X_test['Activity Period'].fillna(X_test['Activity Period'].median())

X_test['Activity Period'].isnull().sum()

X_train['Activity Period'] = pd.to_datetime(X_train['Activity Period'], errors='coerce')

X_train['Activity Period'] = X_train['Activity Period'].apply(lambda x: x.timestamp() if pd.notna(x) else 0)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import Ridge
ridge_model = Ridge(alpha=1.0)
ridge_model.fit(X_train_scaled, Y_train)
Y_pred = ridge_model.predict(X_test_scaled)
```

```python
mse_ridge = mean_squared_error(Y_test, Y_pred)
```

```python
print(f"Ridge Regression Mean Squared Error: {mse_ridge}")
```

```python
from sklearn.linear_model import RidgeCV

r_cv = RidgeCV(alphas=[0.1, 1, 10, 100], store_cv_values=True)
r_cv.fit(X_train, Y_train)

# Best alpha found
print(f"Best alpha: {r_cv.alpha_}")

# Predictions and evaluation
Y_pred = r_cv.predict(X_test)
mse_r_cv = mean_squared_error(Y_test, Y_pred)
print(f"Ridge Regression (CV) Mean Squared Error: {mse_r_cv}")
```

```python
from sklearn.linear_model import Lasso

lasso_model = Lasso(alpha=0.1)
lasso_model.fit(X_train, Y_train)

# Predict on the test set
Y_pred = lasso_model.predict(X_test)

# Evaluate the model (e.g., using mean squared error)
mse_lasso = mean_squared_error(Y_test, Y_pred)

```

```python
print(f"Lasso Regression Mean Squared Error: {mse_lasso}")
```
